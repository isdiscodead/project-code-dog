## 이터레이터 iterator
이터레이터, 반복자는 `반복할 수 있는( iterable )` 객체에서 사용됨

반복할 수 있는 객체 = 반복문의 대상이 될 수 있는 객체 ( 튜플, 리스트, 문자열 등 )

```python
for i in range(1, 10):
    print(i)

type(range(1,10)) # range는 한 번에 생성하지 않아 메모리 절약됨
```

```python
range(1, 10).__iter__() # iter 매직 메서드는 반복자 return
it = iter(range(10)) # 이렇게도 사용 가능 ! 
print( next(it) )
print( next(it) )
print( next(it) )
```

`magic method` = 생성자, iter 메소드처럼 클래스 안에 정의된 특별한 역할을 하는 메소드로 __이름__ 형태



```python
it = iter(range(10))
dir(it)
```

for문은 iterator의 마지막 원소까지 next()를 하면서 진행, 마지막에 탈출하는 원리인 것 !
