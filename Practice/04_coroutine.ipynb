## 코루틴 Coroutine
제네레이터의 또다른 형태로, yield 키워드로 값을 하나씩 반환하는 게 아니라 **값을 외부에서 하나씩 받아서 함수를 실행**함 

```python
def get_three_numbers():
    print("시작")

    given = (yield)
    print(given)

    given = (yield)
    print(given)

    given = (yield)
    print(given)

    print("끝")
```

```python
test = get_three_numbers()
print(test)
```

```python
next(test) # 처음에는 None을 보내거나 next() 사용해야 함 
next(test) # 그냥 next로 호출하면 넘어가는 값이 없어 None이 뜸 
test.send(100)
test.send(500) # 종료되면서 값은 사라짐 
next(test) # 오류 발생
```

```python
def get_three_numbers():
    given = 0
    print("시작")

    given = (yield given)
    print(given)
    given += 1

    given = (yield given)
    print(given)
    given += 1

    given = (yield given)
    print(given)
    given += 1

    yield given
    print("끝")
```
