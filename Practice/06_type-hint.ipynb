## 타입 힌트 type hint

동적 프로그래밍 언어 = 자동으로 type을 유추하여 사용 

c, c++ 등은 명시적으로 타입형을 명시해줘야 하는 정적 프로그래밍 언어

```python
a = 100
type(a)
```

동적 프로그래밍은 편리하지만 코드가 복잡해질수록 혼란을 야기할 수 있음

-> 타입 명시 가능 !! 

```python
number: int = 1
name: str = "홍길동"
arr: list = ['a', 'b', 'c']
```

```python
def circle_width(number: int):
    return number * number * 3.14

circle_width("100") # 잘못된 type을 넣으면 type error 반환 
```
