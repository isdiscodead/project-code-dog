## 함수 형태의 제너레이터
제너레이터( generator ) = 발생자

함수 형태의 이터레이터이다 ! 

```python
def range_1_to_4():
    print('1번째 실행') # yield로 분기를 나눠줌 ! 
    yield 1
    print('2번째 실행')
    yield 2
    print('3번째 실행')
    yield 3
    print('끝')
```

```python
for i in range_1_to_4():
    print(i)

g = range_1_to_4()
print(g)
print(next(g))
print(next(g))
print(next(g))
print(next(g)) # stop iteration error 발생 
```

### 다른 형태의 제너레이터

```python
g = (i for i in range(10))
g
```

```python
g = [i for i in range(10)]
g # 대괄호 사용 시 list 형태 
```

```python
g = (i for i in range(10))

for i in g :
    print(i)
```
